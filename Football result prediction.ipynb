```python
import pandas as pd
import numpy as np
```

```python
matches = pd.read_csv("C:/Users/MadhavaRaj/Documents/Machine Learning/ML Datasets/matches.csv", index_col=0)
matches
```

```python
##Analying each team match count with last two season
##total six teams will be relegated so they play less matches
##find except all teams played all matches 
##one season 38 matches for each team
matches["team"].value_counts()
```

```python
##for one team one matchday count
##1 Matchday = 2
##check postpone matches also
matches["round"].value_counts()
```

```python
##converting home and away as numerical
matches["venue_code"] = matches["venue"].astype("category").cat.codes
##Giving each team an unique opponent number
matches["opp_code"] = matches["opponent"].astype("category").cat.codes
##Converting each match timing as numers - eg:12.30 is converted into 12
matches["hour"] = matches["time"].str.replace(":.+", "", regex=True).astype("int")
```

```python

del matches["comp"]
del matches["notes"]
```

```python
##Giving each day of week a unique number
matches["date"] = pd.to_datetime(matches["date"])
matches["day_code"] = matches["date"].dt.dayofweek
```

```python
matches["target"] = (matches["result"] == "W").astype("int")
```

```python
matches
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf = RandomForestClassifier(n_estimators=50, min_samples_split=10, random_state=1)
```

```python
# important:
# since we are using datetime which is based on timeseries 
# train and test should be given corectly
# for predicting upcoming season train data values may be before the start of new season
# test may be previous season
```

```python
train = matches[matches["date"] < '2022-08-05']
```

```python
test = matches[matches["date"] > '2022-08-05']
```

```python
predictors = ["venue_code", "opp_code", "hour", "day_code"]
```

```python
rf.fit(train[predictors], train["target"])
```

```python
preds = rf.predict(test[predictors])
```

```python
from sklearn.metrics import accuracy_score
error = accuracy_score(test["target"], preds)
error
```

```python
combined = pd.DataFrame(dict(actual=test["target"], predicted=preds))
```

```python
pd.crosstab(index=combined["actual"], columns=combined["predicted"]) 
```

```python
from sklearn.metrics import precision_score
precision_score(test["target"], preds)
```

```python
from sklearn.model_selection import RandomizedSearchCV
# Number of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 200, stop = 1000, num = 10)]
# Number of features to consider at every split
max_features = ['auto', 'sqrt']
# Maximum number of levels in tree
max_depth = [int(x) for x in np.linspace(10, 50, num = 11)]
max_depth.append(None)
# Minimum number of samples required to split a node
min_samples_split = [2, 5, 10]
# Minimum number of samples required at each leaf node
min_samples_leaf = [1, 2, 4]
# Method of selecting samples for training each tree
bootstrap = [True, False]
# Create the random grid
random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf,
               'bootstrap': bootstrap}
print(random_grid)
```

```python
# Use the random grid to search for best hyperparameters
# First create the base model to tune
rf1 = RandomForestClassifier()
# Random search of parameters, using 3 fold cross validation, 
# search across 100 different combinations, and use all available cores
rf1_random = RandomizedSearchCV(estimator = rf, param_distributions = random_grid, n_iter = 30, cv = 2, verbose=2, random_state=42, n_jobs = -1)
# Fit the random search model
rf1_random.fit(train[predictors], train["target"])
```

```python
preds1 = rf1_random.predict(test[predictors])
accuracy_score(test["target"], preds1)
```
